```python
from nltk.tokenize import punkt
```

```python
tokenizer = punkt.PunktSentenceTokenizer() # sentence ko tokenize kiya hai yaha pr
text = 'This is district . This is a state . This is the country .'
sentences = tokenizer.tokenize(text)
sentences
```

```python
## Stemmer ---> word ko root word me karo ,, word ko real or base word me convt krna hai
#Snowball Stemmer
```

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer(language='english')
words = ['running','ran','runs']
stemmed_words = []
for word in words:
  stemmed_word = stemmer.stem(word)
  stemmed_words.append(stemmed_word)

for i in range(len(words)):
  print(f'Original Words: {words[i]} , stemmed_word: {stemmed_words[i]}')
```

```python
# Remove Punctuation from a string
```

```python
import string
text_with_punctuation = "This is @ a string #$ with the punctuation?! ;"
text_without_punctuation = text_with_punctuation.translate(str.maketrans('','',string.punctuation)) # jo punct hai waha pr blank aayega and only words dikhege
print(text_without_punctuation)
```

```python
## POS taggging with nltk
```

```python
from nltk import pos_tag
from nltk import RegexpParser
text = 'Today we will see the demonstration of POS tagging'.split()

print("After split:",text)

tokens_tag = pos_tag(text)
print("After token",tokens_tag)
patterns = """mychunk:{<NN.?>*<VBD.?>*<JJ.?>*<CC>?}"""
chunker = RegexpParser(patterns)
print("After regex:",chunker)
output = chunker.parse(tokens_tag)
print("After Chunking",output)
```
